# Visualization and Plotting

##Correlation Heatmap

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(),annot=True,fmt="1.1f")
```

We have observed from the heatmap that **forestland has a negative correlation**, which has proven to be an important factor, case in point Bulgaria. Also, Savanna fires, forest fires, Fires in organic soils and Fires in humid tropical forests don't have a significant correlation.

---

## Total Emission by years

```python
df.groupby('Year')['total_emission'].sum().plot()
```

The graph depicts the growing trend of CO2 emission all over the globe.

---

## Top 10 Countries by most total emission

```python
df.groupby('Area')['total_emission'].sum().sort_values(ascending=False).head(10).plot.bar()
```

Here we plot the countries with the **highest** total emissions from 1990 to 2020.

---

## Top 10 Countries by least total emission

```python
df.groupby('Area')['total_emission'].sum().sort_values(ascending=True).head(10).plot.bar()
```

Here we plot the countries with the **lowest** total emissions from 1990 to 2020.
